## ¿Qué es convolución?

En palabras sencillas, convolución es aplicar un filtro a todos los píxeles de la imágen, agrupados en pequeños grupos. Se realiza en 3D, pero la idea básica se puede analizar en 2D. 


 









<img src = "https://developer.nvidia.com/sites/default/files/pictures/2018/convolution-1.png" >

## ¿Cómo se aplica matemáticamente? 

Supongamos tenemos una imagen en un solo canal de tamaño $nxn$. Se le aplica un kernel (Una matriz que realiza una convolución). [kernel](https://es.wikipedia.org/wiki/N%C3%BAcleo_(procesamiento_digital_de_im%C3%A1genes))
